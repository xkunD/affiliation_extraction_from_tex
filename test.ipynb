```python
from TexSoup import TexSoup

# Function to extract content between \title and \abstract using TexSoup
def extract_between_title_and_abstract(latex_doc):
    # Parse the document with TexSoup
    soup = TexSoup(latex_doc)
    
    # Convert the parsed document into a list of elements for easy traversal
    elements = list(soup)
    
    # Initialize flags and container for content
    inside_section = False
    between_content = []
    
    # Traverse the elements to find content between \title and \abstract
    for element in elements:
        # Check if the element is the \title
        if element.name == 'title':
            inside_section = True
            continue
        
        # Check if the element is the \abstract, which marks the end of the section
        if element.name == 'abstract':
            inside_section = False
            break
        
        # Collect content between \title and \abstract
        if inside_section:
            between_content.append(str(element))
    
    # Join and clean the extracted content
    extracted_text = ' '.join(between_content).strip()
    cleaned_content = ' '.join(extracted_text.split())  # Removes excess whitespaces/newlines
    
    return cleaned_content if cleaned_content else None

# Sample LaTeX document for testing
latex_doc = """
\\title{Modeling Advection on Directed Graphs using Mat\\'{e}rn Gaussian Processes for Traffic Flow}

\\author{%
  Danielle C. Maddix \\\\
  Amazon Research \\\\
  2795 Augustine Dr. \\\\
  Santa Clara, CA 95054 \\\\
  \\texttt{dmmaddix@amazon.com} \\\\
   \And
   Nadim Saad \\\\
   Stanford University \\\\
   450 Serra Mall \\\\
   Stanford, CA 94305 \\\\
   \\texttt{nsaad31@stanford.edu} \\\\
   \And
   Yuyang Wang \\\\
   Amazon Research \\\\
   2795 Augustine Dr. \\\\
   Santa Clara, CA 95054 \\\\
   \\texttt{yuyawang@amazon.com} \\\\
}

\\abstract{
This is the abstract content of the paper.
}
"""

# Test the function
affiliations_text = extract_between_title_and_abstract(latex_doc)
if affiliations_text:
    print(f"Extracted affiliation content:\n{affiliations_text}")
else:
    print("No affiliation content found between \\title and \\abstract.")
```

```python
import json
import re
import gensim
from TexSoup import TexSoup

# Function to clean and extract content between \name and \begin{abstract}
def extract_between_name_and_abstract(latex_doc):
    match = re.search(r'\\name\{.*?\}\s*(.*?)\\begin\{abstract\}', latex_doc, re.DOTALL)
    
    if match:
        between_content = match.group(1).strip()
        
        # Clean the LaTeX comments and irrelevant sections
        cleaned_content = re.sub(r'\\[A-Za-z]+|%.*|\{.*?\}|\$\^\{\star\}\$', '', between_content)
        cleaned_content = re.sub(r'\s{2,}', ' ', cleaned_content)  # Replace multiple spaces/newlines with a single space
        print('Focus area:', cleaned_content)
        return cleaned_content
    
    return None

# Function to load the research organization registry
def load_registry(ror_path):
    with open(ror_path, 'r') as f:
        registry = json.load(f)
    return registry

# Function to extract institution names from registry
def extract_registry_names(registry):
    institution_names = []
    for entry in registry:
        for name_entry in entry['names']:
            institution_names.append(name_entry['value'])
    return institution_names

# Function to extract affiliations from LaTeX content
def extract_affiliations_from_latex(latex_doc):
    # Extract content between \name and \abstract
    content = extract_between_name_and_abstract(latex_doc)
    
    if content:
        # Split the content by institution separators
        sections = re.split(r',|;', content)
        
        # Clean up the individual sections
        institution_keywords = ['University', 'Institute', 'Research', 'College', 'Laboratory', 'Center', 'School']
        affiliations = [section.strip() for section in sections if any(keyword in section for keyword in institution_keywords)]
        
        print('Affiliations:', affiliations)
        return affiliations if affiliations else []
    
    return []

# Function to find closest institution matches using Gensim
def find_closest_institution(extracted_affiliations, institution_names):
    # Create a Word2Vec model based on the institution names
    institution_tokens = [gensim.utils.simple_preprocess(name) for name in institution_names]
    model = gensim.models.Word2Vec(institution_tokens, vector_size=100, window=5, min_count=1, workers=4)

    closest_matches = []
    
    for affiliation in extracted_affiliations:
        # Preprocess the affiliation for word similarity comparison
        affiliation_tokens = gensim.utils.simple_preprocess(affiliation)
        
        # Ensure that the affiliation tokens are not empty
        if not affiliation_tokens:
            continue  # Skip if the affiliation cannot be tokenized
        
        # Find the closest match for each affiliation in the registry names
        similarities = []
        for name in institution_names:
            name_tokens = gensim.utils.simple_preprocess(name)
            
            # Ensure that both affiliation and name tokens are not empty before comparison
            if not name_tokens:
                continue  # Skip empty institution names
            
            # Calculate similarity
            try:
                similarity = model.wv.n_similarity(affiliation_tokens, name_tokens)
                similarities.append((name, similarity))
            except ZeroDivisionError:
                # Skip comparison if there's an issue with tokenization
                continue
        
        # Find the institution with the highest similarity score
        if similarities:
            best_match = max(similarities, key=lambda x: x[1])
            closest_matches.append((affiliation, best_match))
    
    return closest_matches

# Main function to execute the complete flow
def main(latex_doc_path, registry_json_path):
    # Load the research organization registry
    registry = load_registry(registry_json_path)
    
    # Extract institution names from the registry
    institution_names = extract_registry_names(registry)
    
    # Load and parse LaTeX document
    with open(latex_doc_path, 'r') as f:
        latex_doc = f.read()
    
    # Extract affiliations from LaTeX document
    affiliations = extract_affiliations_from_latex(latex_doc)
    
    # Find closest matches using Gensim
    closest_matches = find_closest_institution(affiliations, institution_names)
    
    # Print the closest matches
    for affiliation, best_match in closest_matches:
        print(f"Extracted affiliation: {affiliation}")
        print(f"Best match: {best_match[0]} (Similarity: {best_match[1]:.2f})")
        print('-' * 40)

# Example usage:
# latex_doc_path = "sample_latex.tex"
# registry_json_path = "ror.json"
# main(latex_doc_path, registry_json_path)
```

```python
import numpy as np
import gensim

def calculate_weighted_similarity(model, tokens1, tokens2, common_terms):
    vec1 = np.zeros(model.vector_size)
    vec2 = np.zeros(model.vector_size)
    
    weight1 = 0
    weight2 = 0
    
    for token in tokens1:
        weight = 0.1 if token in common_terms else 1.0
        vec1 += weight * model.wv[token]
        weight1 += weight
    
    for token in tokens2:
        weight = 0.1 if token in common_terms else 1.0
        vec2 += weight * model.wv[token]
        weight2 += weight
    
    # Normalize vectors
    vec1 /= weight1
    vec2 /= weight2
    
    # Return cosine similarity between the weighted vectors
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

# Function to find closest institution matches using Gensim with weighted similarity
def find_closest_institution_weighted(extracted_affiliations, institution_names, common_terms):
    institution_tokens = [gensim.utils.simple_preprocess(name) for name in institution_names]
    model = gensim.models.Word2Vec(institution_tokens, vector_size=100, window=5, min_count=1, workers=4)

    closest_matches = []
    
    for affiliation in extracted_affiliations:
        affiliation_tokens = gensim.utils.simple_preprocess(affiliation)
        
        if not affiliation_tokens:
            continue
        
        similarities = []
        for name in institution_names:
            name_tokens = gensim.utils.simple_preprocess(name)
            
            if not name_tokens:
                continue
            
            try:
                similarity = calculate_weighted_similarity(model, affiliation_tokens, name_tokens, common_terms)
                similarities.append((name, similarity))
            except KeyError:
                continue
        
        # Find the institution with the highest similarity score
        if similarities:
            best_match = max(similarities, key=lambda x: x[1])
            closest_matches.append((affiliation, best_match))
    
    return closest_matches

common_terms = ['research', 'university', 'institute', 'college', 'laboratory', 'center', 'school']
extracted_affiliations = ['Amazon Research', 'Stanford University', 'Amazon Research']
institution_names = ['Amazon.com', 'Stanford University', 'Rainbow Research', 'MIT']

closest_matches = find_closest_institution_weighted(extracted_affiliations, institution_names, common_terms)

for affiliation, best_match in closest_matches:
    print(f"Extracted affiliation: {affiliation}")
    print(f"Best match: {best_match[0]} (Similarity: {best_match[1]:.2f})")
    print('-' * 40)
```

```python
import os
import json
import numpy as np
import gensim
from glob import glob

# Function to remove department part and get the institution name
def extract_institution_from_affiliation(affiliation):
    # Split by the first comma and take the part after it, or use the whole string if no comma exists
    return affiliation.split(',', 1)[-1].strip()

# Function to calculate weighted similarity between two sets of tokens
def calculate_weighted_similarity(model, tokens1, tokens2, common_terms):
    vec1 = np.zeros(model.vector_size)
    vec2 = np.zeros(model.vector_size)
    
    weight1 = 0
    weight2 = 0
    
    for token in tokens1:
        weight = 0.1 if token in common_terms else 1.0
        vec1 += weight * model.wv[token]
        weight1 += weight
    
    for token in tokens2:
        weight = 0.1 if token in common_terms else 1.0
        vec2 += weight * model.wv[token]
        weight2 += weight
    
    # Normalize vectors
    vec1 /= weight1
    vec2 /= weight2
    
    # Return cosine similarity between the weighted vectors
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

# Function to find closest institution matches using Gensim with weighted similarity
def find_closest_institution_weighted(extracted_affiliations, institution_names, common_terms):
    institution_tokens = [gensim.utils.simple_preprocess(name) for name in institution_names]
    model = gensim.models.Word2Vec(institution_tokens, vector_size=100, window=5, min_count=1, workers=4)

    closest_matches = []
    
    for affiliation in extracted_affiliations:
        # Extract the institution part after the first comma
        affiliation_institution = extract_institution_from_affiliation(affiliation)
        
        # Tokenize the affiliation
        affiliation_tokens = gensim.utils.simple_preprocess(affiliation_institution)
        
        if not affiliation_tokens:
            continue
        
        similarities = []
        for name in institution_names:
            name_tokens = gensim.utils.simple_preprocess(name)
            
            if not name_tokens:
                continue
            
            try:
                similarity = calculate_weighted_similarity(model, affiliation_tokens, name_tokens, common_terms)
                similarities.append((name, similarity))
            except KeyError:
                continue
        
        # Find the institution with the highest similarity score
        if similarities:
            best_match = max(similarities, key=lambda x: x[1])
            closest_matches.append((affiliation, best_match))
    
    return closest_matches

# Main function to process JSON files and perform matching
def process_json_files(json_folder, ror_data_path, common_terms):
    # Load the institution names from ror.json
    with open(ror_data_path, 'r') as ror_file:
        ror_data = json.load(ror_file)
    
    # Extract institution names from ror.json
    institution_names = [name['value'] for institution in ror_data for name in institution['names']]

    # Loop through each JSON file in the folder
    json_files = glob(os.path.join(json_folder, '*.json'))
    
    for json_file in json_files:
        with open(json_file, 'r') as f:
            authors_data = json.load(f)
        
        # Extract affiliations from each author
        extracted_affiliations = []
        for author, details in authors_data.items():
            affiliations = details.get('Affiliation', [])
            extracted_affiliations.extend(affiliations)
        
        # Perform approximate matching of affiliations to ROR institution names
        closest_matches = find_closest_institution_weighted(extracted_affiliations, institution_names, common_terms)
        
        # Print the results for each file
        print(f"Results for file: {json_file}")
        for affiliation, best_match in closest_matches:
            print(f"Extracted affiliation: {affiliation}")
            print(f"Best match: {best_match[0]} (Similarity: {best_match[1]:.2f})")
            print('-' * 40)

# Define the common terms to de-weight in similarity calculations
common_terms = ['research', 'university', 'institute', 'college', 'laboratory', 'center', 'school']

# Define the folder containing the author JSON files and the ROR data file
json_folder = './2201_samp_test_set/'
ror_data_path = 'ror.json'

# Run the processing function
process_json_files(json_folder, ror_data_path, common_terms)
```
